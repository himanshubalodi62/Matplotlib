# Plotting x and y points

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

```python
#let's this is dataset create different.. plot.
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

Labels for a Plot

With Pyplot use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

Title for a Plot

With Pyplot, you can use the title() function to set a title for the plot.

# Linestyle

You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=5, linestyle='dotted')
```

# Shorter Syntax

The line style can be written in a shorter syntax:

linestyle can be written as ls.

dotted can be written as :.

dashed can be written as --.

# plot format strings and other options....

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'b+--')

```

# Markers

You can use the keyword argument marker to emphasize each point with a specified marker:

```python
# Same effect as 'b+' format string
plt.plot(x,y,color='blue',marker='+',linestyle='')
```

# markersize..

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```

# alpha property to control transparency of line chart

```python
plt.plot(x,y,'g<',alpha=0.5) # alpha can be specified on a scale 0 to 1
```

# .............................Bar Chart........................................

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with

heights or lengths proportional to the values that they represent. The bars can be plotted vertically

or horizontally.

The bar() function takes arguments that describes the layout of the bars.

The categories and their values represented by the first and second argument as arrays.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple bar chart showing revenues of major US tech companies....
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
ypos = np.arange(len(company)) #here store the string 
ypos
```

```python
xpos = np.arange(len(company))
xpos   #its replace the numpy values as company..
```

```python

plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

# Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

# Horizontal bar chart using barh function

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

# .......................... Scatter Plots........................

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

# .......................................Histogram.................................

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

In histograms, x axis contains a variable and y axis will be a frequency of that variable

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

# example:
We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic



```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
```

# bins parameter...

The bins are usually specified as consecutive, non-overlapping intervals of a variable.

The matplotlib.pyplot.hist() function plots a histogram. It computes and draws the histogram of x.

# hist function :
The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

# Histogram showing normal, pre-diabetic and diabetic patients distribution

. 80-100: Normal

. 100-125: Pre-diabetic

. 80-100: Diabetic

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")


plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

# Mutiple data samples in a histogram


```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()

```

# histtype=step

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step')

```

# horizontal orientation

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
```

# ...........................Pie charts...........................................

A Pie Chart can only display one series of data. Pie charts show the size of items (called wedge) in

one data series, proportional to the sum of the items. The data points in a pie chart are shown as a

percentage of the whole pie.


# Parameters...

x array-like. The wedge sizes.
# labels :
list. A sequence of strings providing the labels for each wedge.
# Colors: 
A sequence of matplotlibcolorargs through which the pie chart will cycle. If None, will use the colors in the currently active cycle.
# Autopct:
string, used to label the wedges with their numeric value. The label will be placed inside the wedge. The format string will be fmt%pct.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

Draw pie chart to track down home expenses

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

 # Draw pie chart as perfect circle

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

# Show percentages for every pie. Specify radius to increase chart size

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5)
plt.show()
```

# Explode:

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```

# Saving chart to a file using savefig

# ..........................Box plot............................................

A box plot which is also known as a whisker plot displays a summary of a set of data containing the

minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the

first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers

go from each quartile to the minimum or maximum.

Box Plot

We use the numpy.random.normal() function to create the fake data. It takes three arguments, mean and

standard deviation of the normal distribution, and the number of values desired.

```python
df = (100, 10, 200,50,46,3434343,23)
```

```python
fig = plt.figure()
# Create an axes instance
ax = fig.add_axes([0,0,1,1])
# Create the boxplot
bp = ax.boxplot(df)
plt.show()
```
